### Linearne jednadžbe

Probajmo pomoću korištenja inverza matrica riješiti sustav jednadžbi.

Proizvodimo kekse, mafine i torte. Svaki dan nabavimo 100 jaja, 150 šalica jaja i moramo ukupno proizvesti 100 proizvoda (kekse, mafine i torte).
Za jedan keks treba nam 1 šalica brašna i 1 jaje.
Za jedan mafin treba nam 2 šalice brašna i 1 jaje.
Za tortu treba nam 3 šalice brašna i 2 jaja.
Pitanje je, koliko točno svakog od proizvoda moramo napraviti, ali tako da potrošimo sve što smo nabavili za taj dan.

```python
A = np.matrix([[1, 1, 1], [1, 1, 2], [1, 2, 3]])
A
```

```python
b = np.array([100, 100, 150])
b = np.matrix(b)
```

```python
x = np.linalg.inv(A) @ b.transpose()  # Zasto koristimo transponse?
x
```

NumPy pruža i funkciju za rješavanje linearnih jednadžbi.
Više informacija: https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html

```python
x = np.linalg.solve(A, b.transpose())
x
```

### Pseudoinverz matrice

Pogledajmo na konkretnom primjeru modeliranja podataka kako iskoristiti pseudoinverz matrice.

Učitajmo podatke o automobilima i to snagu i težinu:

```python
import pandas as pd

automobili = pd.read_csv('cars.csv', usecols=['Horsepower', 'Weight'], delimiter=';')
automobili.drop(0, inplace=True)
automobili = automobili.astype(float)
automobili
```

Probajmo to i prikazati u grafu.

```python
automobili.plot.scatter(x='Horsepower', y='Weight')
```

Želimo iscrtati pravac koji najbolje opisuje ove podatke. Vidimo da imamo i snagu postavljenu u 0. To je očito pogrešan podatak, pa ćemo sve takve automobile izbrisati iz skupa.

```python
automobili = automobili[automobili['Horsepower'] != 0]
automobili.plot.scatter(x='Horsepower', y='Weight')
```

Izračunat ćemo parametre pravca pomoću pseudoinverza matrica.

```python
A = np.matrix(automobili['Horsepower']).T # DataFrame pretvaramo u matricu!
A
```

```python
b = np.matrix(automobili[['Weight']])
b
```

```python
H = np.hstack((A, np.ones((A.shape[0], 1))))
x = np.linalg.inv(H.T * H) * H.T * b
x
```

```python
import matplotlib.pyplot as plt

def pravac(x_tocka):
    return float(np.matrix([x_tocka, 1])* x)

x1 = 0
x2 = 300
y1 = pravac(x1)
y2 = pravac(x2)

plt.plot([x1, x2], [y1, y2], '-r')
plt.scatter(x=automobili[['Horsepower']], y=automobili[['Weight']])
```

Naravno, NumPy implementira i funkciju koja sve ovo može napraviti za nas.
https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html

Poslužit ćemo se i s metodom `flatten`: https://numpy.org/doc/stable/reference/generated/numpy.ndarray.flatten.html

Usporedite rezultate.

```python
x = automobili[['Horsepower']].to_numpy().flatten()
y = automobili[['Weight']].to_numpy().flatten()
k, b = np.polyfit(x, y, 1)

print(k, b)

x1 = 0
x2 = 300
y1 = x1*k + b
y2 = x2*k + b

plt.plot([x1, x2], [y1, y2], '-r')
plt.scatter(x=automobili[['Horsepower']], y=automobili[['Weight']])
```
