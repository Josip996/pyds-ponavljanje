Odgovoriti na pitanje koristeci Pandas, NumPy i Matplotlib ili Pandas.plot() funkciju

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
penguins = pd.read_csv('data/penguins.csv')
```

```python
penguins.head(5)
penguins.tail(5)
```

```python
penguins.drop(columns=['island', 'sex', 'year'], inplace=True)
groupby_species = penguins.groupby(by=['species']).mean()
```

```python
groupby_species
```

```python
groupby_species.plot(kind='bar', figsize=(10, 6))
```
