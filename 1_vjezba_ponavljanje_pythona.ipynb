## Grananje

Ako varijabla moze imati vise vrijednosti onda koristimo grananje.
Primjer:
Ako lista ima dva elementa ispisati "Mala lista". Ako lista ima vise od dva elementa i manje i jednako od 5, ispisati "Srednja lista". I ako je lista veca od 5 ispisati "Lista je velika".

```python
lista = [1, 2, 3, 4, 5, 6]

if len(lista) <= 2:
    print('Mala lista')
elif len(lista) > 2 and len(lista) <= 5:
    print('Srednja lista')
else:
    print('Velika lista')
```

```python
# za svaki ELEMENT u KOLEKCIJI:
#    pokreni neku aktivnost
# for ELEMENT in COLLECTION:
#    start action
for i in range(11):
    print(i, end=', ')
```

```python
numbers = [1, 2, 3, 4, 5, 6]

for number in numbers:
    if number == numbers[-1]:
        print(number, end='')
    else:
        print(number, end=', ')
```

```python
for index, element in enumerate(numbers):
    if index == len(numbers) - 1:
        print(element, end='')
    else:
        print(element, end=', ')
```

```python
# while uvjet:
#    kod koji se izvrsava

i = 1
while i < 10:
    if i == 9:
        print(i, end='')
    else:
        print(i, end=', ')
    # i = i + 1
    i += 1
```

```python
number = 10
while True:
    if number == 0:
        break
    else:
        number -= 1
        print(number, end=' ')


# nakon break, prekida se iyvrsavanje petlje i nastavalja se izvrsvat kod, od ove linije na dalje.
```

```python
number = 10
while True:
    if number == 0:
        number -= 1
        continue
    elif number == -10:
        break
    else:
        print(number, end=' ')
        number -= 1

# Nakon continue prekida se izvrsavanje ciklusa petlje, 
# ali se petlja i dalje nastavlja izvresavati
```

Nakon toga trebamo kreirati funkciju koja na osnovu proizvoda (rjecnik) generira stavku racuna (takoder rjecnik). Ideja je da se iz jednog rjecnika dohvate odredeni podaci i prosire novima kao sto je kolicina i ukupan iznos.

```python
def get_invoice_item(product, amount):
    if amount <= 0:
        return {}
    return {
        'title': f"{product['title']} ({product['code']})",
        'amount': amount,
        'unit': 'kom.',
        'unit_price': product['price'],
        'total_price': product['price'] * amount
    }
```

Sada nam treba funkcija koja generira racun na osnovu liste stavki.
U ovoj funkciji je u liniji 5 kraci oblik for petlje koja prolazi kroz listu i iz svakog elementa liste (rjecnik) uzima vrijednost koja se nalazi pod kljucem `'total_price'` i od toga radi novu listu (one uglate zagrade), nakon toga pomocu funkcije sum() se zbrajaju svi elementi te nove liste i ta vrijednost se pridruzuje ukupnom iznosu bez PDVa

```python
def get_invoice(invoice_items):
    # total_wo_vat = 0
    # for invoice_item in invoice_items:
    #     total_wo_vat += invoice_item['total_price']
    total_wo_vat = sum([invoice_item['total_price'] for invoice_item in invoice_items])
    vat = total_wo_vat * 0.25
    total = total_wo_vat + vat
    return {
        'invoice_items': invoice_items,
        'total_with_out_vat': total_wo_vat,
        'vat': vat,
        'total': total
    }
```
