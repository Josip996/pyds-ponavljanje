```python
tekst = 'Ovo je tekst koji zelim upisati u datoteku.'

datoteka = open('demo.txt', 'w')
datoteka.write(tekst)
datoteka.close()
```

```python
tekst = 'Ovo je nastavak teksta koji zelim upisati u datoteku.'

try:
    datoteka = open('demo.txt', 'w')
    datoteka.write(tekst)
except Exception as ex:
    print(f'Dogodila se greska: {ex}')
finally:
    datoteka.close()
```

# Klase i objekti



```python
class Employee:
    def __init__(self, first_name: str, last_name: str, age: int, job_position:str ='Djelatnik'):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.job_position = job_position

    def is_mature(self):
        if self.age >= 18:
            return True
        else:
            return False

    def is_older_then(self, other_person: Employee):
        return self.age > other_person.age

    def __str__(self):
        return f'{self.first_name} {self.last_name} ({self.age}) - {self.job_position}'


class Student:
    def __init__(self, first_name: str, last_name: str, age: int):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def is_mature(self):
        if self.age >= 18:
            return True
        else:
            return False

    def is_older_then(self, other_person: Employee):
        return self.age > other_person.age

    def __str__(self):
        return f'{self.first_name} {self.last_name} ({self.age})'















pero = Employee('Pero', 'Peric', 25, 'Senior Python Developer')
iva = Employee('Iva', 'Ivic', 30, 'CEO')
marko = Employee('Marko', 'Maric', 28)
leo = Student('Leo', 'Leic', 26)


print(pero)
print(iva)
print(marko)

if iva.is_mature():
    print(f'{iva.first_name} je punoljetna.')
else:
    print(f'{iva.first_name} NIJE punoljetna.')

if iva.is_older_then(pero):
    print(f'{iva.first_name} je starija od {pero.first_name}.')
else:
    print(f'{iva.first_name} NIJE starija od {pero.first_name}.')


if leo.is_mature():
    print(f'{leo.first_name} je punoljetan.')
else:
    print(f'{leo.first_name} NIJE punoljetan.')

if iva.is_older_then(leo):
    print(f'{iva.first_name} je starija od {leo.first_name}.')
else:
    print(f'{iva.first_name} NIJE starija od {leo.first_name}.')
```

```python
class Person:
    def __init__(self, first_name: str, last_name: str, age: int, date_of_birth: str = '1900-01-01'):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.date_of_birth = date_of_birth

    def is_mature(self):
        if self.age >= 18:
            return True
        else:
            return False

    def is_older_then(self, other_person: Person):
        return self.age > other_person.age

    def __str__(self):
        return f'{self.first_name} {self.last_name} ({self.age})'

    

class Employee(Person):
    def __init__(self, first_name: str, last_name: str, age: int, job_position:str ='Djelatnik', date_of_birth: str = '1900-01-01'):
        super().__init__(first_name, last_name, age, date_of_birth)
        self.job_position = job_position

    def __str__(self):
        return f'{self.first_name} {self.last_name} ({self.age}) - {self.job_position}'


class Student(Person):
    def __init__(self, first_name: str, last_name: str, age: int, date_of_birth: str = '1900-01-01'):
        super().__init__(first_name, last_name, age, date_of_birth)













pero = Employee('Pero', 'Peric', 25, 'Senior Python Developer')
iva = Employee('Iva', 'Ivic', 30, 'CEO')
marko = Employee('Marko', 'Maric', 28)
leo = Student('Leo', 'Leic', 26, '2000-01-01')


print(pero)
print(iva)
print(marko)

if iva.is_mature():
    print(f'{iva.first_name} je punoljetna.')
else:
    print(f'{iva.first_name} NIJE punoljetna.')

if iva.is_older_then(pero):
    print(f'{iva.first_name} je starija od {pero.first_name}.')
else:
    print(f'{iva.first_name} NIJE starija od {pero.first_name}.')


if leo.is_mature():
    print(f'{leo.first_name} je punoljetan.')
else:
    print(f'{leo.first_name} NIJE punoljetan.')

if iva.is_older_then(leo):
    print(f'{iva.first_name} je starija od {leo.first_name}.')
else:
    print(f'{iva.first_name} NIJE starija od {leo.first_name}.')


print(iva.date_of_birth)
print(leo.date_of_birth)
```
